```python
data = pd.DataFrame([ {"n_chambres" : 2, "prix" : 10}, {"n_chambres" : 4, "prix" : 20},
                    {"n_chambres" : 3, "prix" : 15},
                     {"n_chambres" : 3, "prix" : 20},
                     {"n_chambres" : 1, "prix" : 3.5},
                    {"n_chambres" : 6, "prix" : 31},
                     {"n_chambres" : 5, "prix" : 33},
                    {"n_chambres" : 8, "prix" : 36},
                    {"n_chambres" : 7, "prix" : 50},
                    {"n_chambres" : 9, "prix" : 42},
                    ])
data
```

```python
plt.scatter(data['n_chambres'], data['prix'], c="r") #Pour afficher les points
#plt.plot(data['n_chambres'], data['prix'], c="b") #Pour afficher la droite de rgression
plt.xlabel("Nombre de chambre") # Axe des X (repsantant le nombe de chambre)
plt.ylabel("Prix") # Axe des Y reprsantant le prix
plt.title("Prix en fonction du nombre de chambre") # Titre
plt.show()
```

```python
plt.scatter(data['n_chambres'], data['prix'], color='red')
plt.plot(data['n_chambres'], data['n_chambres'] * 5, color='blue', label="Modle : y=5x")
plt.xlabel("Nombre de chambres")
plt.ylabel("Prix (en milliers)")
plt.title("Distribution des donnes")
plt.show()
```

```python
def model(x, w):
    return w * x
```

```python
# Tracer les donnes
plt.scatter(data['n_chambres'], data['prix'], color='red', label="Donnes")

# Essayer diffrentes valeurs de w
for w in [3, 4, 5, 6]:
    plt.plot(data['n_chambres'], model(data['n_chambres'], w), label=f"w = {w}")

plt.xlabel("Nombre de chambres")
plt.ylabel("Prix (en milliers)")
plt.title("Diffrentes droites d'ajustement")
plt.legend()
plt.show()
```

```python
def erreur(x, w, y):
    # Calculer les prdictions
    y_model = model(x, w)
    # Calculer la distance (erreur quadratique)
    distance = (y - y_model) ** 2
    # Retourner la somme des distances
    return np.sum(distance)
```

```python
# Tester diffrentes valeurs de w
for w in [3, 4, 5, 6]:
    err = erreur(data['n_chambres'], w, data['prix'])
    print(f"Erreur pour w = {w}: {err}")
```

```python
for w in [4.8, 4.9, 5, 5.1]:
    err = erreur(data['n_chambres'], w, data['prix'])
    print(f"Erreur pour w = {w}: {err}")
```

```python
possible_w = np.arange(-20, 20, 0.001)
possible_w
```

```python
errors = []

for w in possible_w:
    e = erreur(x=data['n_chambres'], w=w, y=data['prix'])
    errors.append(e)
```

```python
plt.plot(possible_w, errors)
plt.xlabel('Valeurs possibles de w')
plt.ylabel('Erreur totale')
plt.title('Erreur en fonction des valeurs de w')
plt.show()
```

```python
min_error = np.min(errors)
print("Erreur minimale:", min_error)
optimal_index = np.argmin(errors)
print("Index optimal:", optimal_index)
optimal_w = possible_w[optimal_index]
print("Valeur optimale de w:", optimal_w)
```
