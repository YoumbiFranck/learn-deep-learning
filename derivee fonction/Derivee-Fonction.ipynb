```python
# Définir la fonction f(x) = x²
def f(x):
    return x**2

# Définir une fonction pour calculer la dérivée approximative
def derivee(fonction, point, delta=0.0001):
    """
    Calcule la dérivée approximative d'une fonction au point donné.

    Arguments :
    - fonction : La fonction dont on calcule la dérivée.
    - point : Le point où la dérivée est calculée.
    - delta : Le petit changement appliqué à x (par défaut 0.0001).

    Retourne :
    - Le taux de variation (la dérivée approximative).
    """
    variation = fonction(point + delta) - fonction(point)
    taux_variation = variation / delta
    return taux_variation

# Calculer la dérivée de f(x) en 3
point = 3
resultat = derivee(f, point)

print(f"La dérivée approximative de f(x) = x² au point x={point} est {resultat}")

# Calculer la dérivée pour un autre point
point_negatif = -5
resultat_negatif = derivee(f, point_negatif)
print(f"La dérivée approximative de f(x) = x² au point x={point_negatif} est {resultat_negatif}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Visualiser f(x) et sa tangente en x=3
x_vals = np.linspace(-6, 6, 500)  # Intervalle pour dessiner la courbe
y_vals = f(x_vals)

# Calculer la pente (dérivée) au point x=3
pente = derivee(f, point)
tangente = lambda x: pente * (x - point) + f(point)  # Équation de la tangente

# Graphique
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = x^2$", color="blue")
plt.scatter([point], [f(point)], color="red", label=f"Point étudié (x={point}, f(x)={f(point)})")
plt.plot(x_vals, tangente(x_vals), '--', label=f"Tangente en x={point}", color="orange")

plt.title("Illustration de la dérivée comme pente de la tangente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
plt.axvline(0, color="black", linewidth=0.7, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```
